```python
import random

def isPrime(number):
    if ( (2 ** (number -1) ) % number ) == 1:
        return (True ) 
    else:
        return False


def gen_prime(max):
    for i in range(max,2,-1):
        if(isPrime(i)):
            return i
        
        
def gcd(a, b):
    if b == 0:
        return (a, 1, 0)
    (d_, x_, y_) = gcd(b, a % b)
    (d, x, y) = (d_, y_, x_ - (a // b) * y_)
    return (d, x, y)

        
def gen_privkey(p, q, e):
    f_n = (p - 1) * (q - 1)
    (x, d, y) = gcd(e, f_n)
    if d < 0: d += f_n
    return d


def pubkey_encrypt(text, pubkey, factor):
    return text ** pubkey % factor


def privkey_decrypt(chiffre, privkey, factor):
    return chiffre **  privkey % factor

print ("Генерация простого числа p")
prime_p = gen_prime(random.randint(100,500))
print ("p = " + str(prime_p))

print ("Генерация простого числа q")
prime_q = gen_prime(random.randint(100,500))
print ("q = " + str(prime_q))

assert prime_p != prime_q

print ("Вычисление n=p*q")
factor_n = prime_p*prime_q
print ("n = " + str(factor_n))

print ("Вычисление f_n")
f_n = (prime_p-1)*(prime_q-1)
print ("f:" + str(f_n))


e = 17
print ("Значение закрытого ключа e:" + str(e))

assert e < f_n

print ("Генерация ключа d")
privkey_d = gen_privkey(prime_p, prime_q, e)
print ("d:" + str(privkey_d))
print('============')
def my_rsa(text):
    print ("Исходное сообщение = " + str(text))
    text_encrypted = pubkey_encrypt(text, e, factor_n)
    print ("Зашифрованное сообщение: " + str(text_encrypted))
    text_decrypted = privkey_decrypt(text_encrypted, privkey_d, factor_n)
    print ("Расшифрованное сообщение: " + str(text_decrypted))

my_rsa(text = 8)
print('============')
my_rsa(text = 27)
print('============')
my_rsa(text = 5)
print('============')
my_rsa(text = random.randint(0,100))
```
